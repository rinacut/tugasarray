```python
import matplotlib.pyplot as plt

# Data contoh
usia = [20, 22, 21, 23, 20, 22, 21, 24]
ipk = [3.5, 3.2, 3.8, 2.9, 3.1, 3.6, 2.7, 3.9]

# grafiknya 
plt.figure(figsize=(8,4))
plt.bar(usia, ipk, color='b')
plt.title('Bar Chart Usia dan Ipk')
plt.xlabel('Usia')
plt.ylabel('Ipk')
plt.grid(False)
plt.show()
```
